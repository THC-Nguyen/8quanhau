```python

import tkinter as tk

BOARD_SIZE = 8

# Vị trí 8 quân hậu (dạng [row, col], bắt đầu từ 0)
queen_positions = [
    (0, 3), (1, 6), (2, 2), (3, 7),
    (4, 1), (5, 4), (6, 0), (7, 5)
]

CELL_SIZE = 60  # kích thước mỗi ô cờ


def draw_board(canvas, with_queens=False):
    for row in range(BOARD_SIZE):
        for col in range(BOARD_SIZE):
            x1, y1 = col * CELL_SIZE, row * CELL_SIZE
            x2, y2 = x1 + CELL_SIZE, y1 + CELL_SIZE
            color = "white" if (row + col) % 2 == 0 else "gray"
            canvas.create_rectangle(x1, y1, x2, y2, fill=color, outline="black")

            if with_queens and (row, col) in queen_positions:
                canvas.create_text(
                    x1 + CELL_SIZE // 2,
                    y1 + CELL_SIZE // 2,
                    text="♛",
                    font=("Arial", 28),
                    fill="red"
                )


def main():
    root = tk.Tk()
    root.title("Bài toán 8 quân hậu")

    # Khung chứa 2 bàn cờ
    frame = tk.Frame(root)
    frame.pack(padx=50, pady=50)

    # Khung điều khiển (nút + đồng hồ)
    control_frame = tk.Frame(root)
    control_frame.pack(pady=20)

    # Nhãn hiển thị thời gian
    timer_label = tk.Label(control_frame, text="Thời gian: 0 giây", font=("Arial", 14))
    timer_label.grid(row=0, column=0, columnspan=3, pady=5)

    # Các nút điều khiển
    btn_start = tk.Button(control_frame, text="Bắt đầu")
    btn_start.grid(row=1, column=0, padx=5)

    btn_pause = tk.Button(control_frame, text="Tạm dừng")
    btn_pause.grid(row=1, column=1, padx=5)

    btn_reset = tk.Button(control_frame, text="Top (Reset)")
    btn_reset.grid(row=1, column=2, padx=5)

    # Bàn cờ bên trái (trống)
    canvas_left = tk.Canvas(frame, width=BOARD_SIZE * CELL_SIZE, height=BOARD_SIZE * CELL_SIZE)
    canvas_left.grid(row=0, column=0, padx=50)
    draw_board(canvas_left, with_queens=False)

    # Bàn cờ bên phải (có quân hậu)
    canvas_right = tk.Canvas(frame, width=BOARD_SIZE * CELL_SIZE, height=BOARD_SIZE * CELL_SIZE)
    canvas_right.grid(row=0, column=1, padx=50)
    draw_board(canvas_right, with_queens=True)

    root.mainloop()


if __name__ == "__main__":
    main()
```
